# Solar Electricity

![](data/007_q0.jpg "The outside surface of the NASA IMAGE satellite is covered with solar cells to collect sunlight and generate electricity for its many instruments and systems.")

Most satellites depend on sunlight converted into electricity and stored in batteries, to run their instruments. Typical research satellites operate with only **200 to 800 watts** of electricity generated by sunlight and _solar cells_. 

_Solar cells_ can be attached directly to the outer surface of a satellite, or can be found on _solar panels_ that the satellite deploys after it reaches its orbit. Engineers decide how many solar cells are needed by finding out how much power the satellite needs to operate, and then figuring out what the satellite's surface area is. If the satellite is not big enough, additional solar panels may be needed to supply the electricity.

In this exercise, you will calculate the perimeter of the satellite, in centimeters, and the area, in square centimeters, of several kinds of satellite surfaces. You will then decide if there is enough available surface to supply the electrical needs of the satellite. 

All measurements in the diagrams are in centimeters. Some dimensions may be missing.

### Question 1

Use the figure below to answer the following questions: 

* What is the perimeter?
* What is the surface area? 

![](data/007_q1.JPG)

The solar cells can provide **0.03 watts per square centimeter**, and the satellite needs **257 watts**. 

* Is there enough surface area in the figure to meet the electrical needs?

### Answer 1

* The missing side can be reconstructed from the other measurements, as calculated below. (You could also have deduced that the perimeter would be the same as a square with four sides totalling 120 centimeters each!)

```python
# Total perimeter of Figure 1, in centimeters.
q1_perimeter = 2 * (120 + 50 + 30 + 40)
q1_perimeter
```

* The area can be found by breaking the figure into rectagular sections, from right to left.

```python
# Total surface area of Figure 1, in square centimeters.
q1_area = (120 * 50) + (30 * 70) + (40 * 40)
q1_area
```

* The solar cells produce **0.03 watts per square cm**, so the power available is:

```python
# Power available to the satellite, in watts.
q1_area * 0.03
```

This is greater than the minimum satellite requirement of **257 watts**, so no additional solar panels will need to be added. Our spacecraft is heading to orbit!

* The surface area can be found from breaking the figure into a rectangle and a semi-circle.

```python
# Surface area of Figure 2, in square centimeters.
q2_area = (135 * q2_rectangle_length) + 0.5 * math.pi * q2_radius**2
q2_area
```

* The solar cells produce **0.03 watts per square centimeter**, so the available power is:

```python
# Power available to the satellite, in watts.
q2_area * 0.03
```

This is greater than the minimum satellite requirement of **957 watts**, so no additional solar panels will need to be added. Our spacecraft is heading to orbit!
